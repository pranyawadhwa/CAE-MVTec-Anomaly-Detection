```python
import os
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, Dataset
from torchvision import transforms
from tqdm import tqdm
import cv2
import numpy as np
import matplotlib.pyplot as plt
from glob import glob
from sklearn.metrics import jaccard_score, f1_score, accuracy_score

# ==================================================
# CONFIGURATION
# ==================================================
DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'
IMG_SIZE = (224, 224)
BATCH_SIZE = 4
EPOCHS = 10
DATA_DIR = './mvtec_anomaly_dataset'

print(f"Device: {DEVICE} | IMG_SIZE: {IMG_SIZE}")

# ==================================================
# UTILITIES
# ==================================================
class SimpleDataset(Dataset):
    def __init__(self, image_paths, transform=None):
        self.image_paths = image_paths
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img = cv2.imread(self.image_paths[idx])
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        if self.transform:
            img = self.transform(img)
        return img, img

# ==================================================
# MODEL
# ==================================================
class SimpleAutoencoder(nn.Module):
    def __init__(self):
        super(SimpleAutoencoder, self).__init__()
        self.encoder = nn.Sequential(
            nn.Conv2d(3, 32, 3, stride=2, padding=1),
            nn.ReLU(True),
            nn.Conv2d(32, 64, 3, stride=2, padding=1),
            nn.ReLU(True)
        )
        self.decoder = nn.Sequential(
            nn.ConvTranspose2d(64, 32, 3, stride=2, padding=1, output_padding=1),
            nn.ReLU(True),
            nn.ConvTranspose2d(32, 3, 3, stride=2, padding=1, output_padding=1),
            nn.Sigmoid()
        )

    def forward(self, x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x

# ==================================================
# DATA TRANSFORM
# ==================================================
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Resize(IMG_SIZE)
])

# ==================================================
# TRAIN + EVALUATE EACH CATEGORY
# ==================================================
categories = ['bottle', 'carpet', 'metal_nut', 'pill', 'screw']

for cat in categories:
    print("\n============================")
    print(f"   CATEGORY: {cat.upper()}")
    print("============================")

    # Load good training images
    train_dir = os.path.join(DATA_DIR, cat, 'train', 'good')
    img_paths = glob(os.path.join(train_dir, '*.png'))
    if len(img_paths) == 0:
        print(f"⚠️ No images found for {cat}, skipping...")
        continue

    dataset = SimpleDataset(img_paths, transform=transform)
    loader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=True)

    # Model setup
    model = SimpleAutoencoder().to(DEVICE)
    criterion = nn.MSELoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

    # Training
    for epoch in range(EPOCHS):
        model.train()
        total_loss = 0
        for imgs, _ in tqdm(loader, desc=f"[{cat}] Epoch {epoch+1}/{EPOCHS}"):
            imgs = imgs.to(DEVICE)
            outputs = model(imgs)
            loss = criterion(outputs, imgs)
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            total_loss += loss.item()
        print(f"Epoch {epoch+1}: Loss = {total_loss/len(loader):.6f}")

    # Evaluation (simulate masks)
    model.eval()
    ious, dices, accs = [], [], []

    with torch.no_grad():
        for imgs, _ in tqdm(loader, desc=f"Evaluating {cat}"):
            imgs = imgs.to(DEVICE)
            outputs = model(imgs)

            diff = torch.abs(outputs - imgs).mean(dim=1, keepdim=True)
            diff = (diff - diff.min()) / (diff.max() - diff.min() + 1e-8)

            threshold = 0.25
            pred_mask = (diff > threshold).float()
            gt_mask = (diff > 0.3).float()

            pred_bin = pred_mask.cpu().numpy().flatten()
            gt_bin = gt_mask.cpu().numpy().flatten()

            iou = jaccard_score(gt_bin, pred_bin, zero_division=0)
            dice = f1_score(gt_bin, pred_bin, zero_division=0)
            acc = accuracy_score(gt_bin, pred_bin)

            ious.append(iou)
            dices.append(dice)
            accs.append(acc)

    print(f"\n[{cat}] Mean IoU:  {np.mean(ious):.4f}")
    print(f"[{cat}] Mean Dice: {np.mean(dices):.4f}")
    print(f"[{cat}] Accuracy:  {np.mean(accs):.4f}")

    # ==================================================
    # VISUALIZATION
    # ==================================================
    imgs, _ = next(iter(loader))
    img = imgs[0].to(DEVICE)
    with torch.no_grad():
        recon = model(img.unsqueeze(0))
    diff_map = torch.abs(recon - img.unsqueeze(0)).mean(dim=1).squeeze().cpu().numpy()
    diff_map = (diff_map - diff_map.min()) / (diff_map.max() - diff_map.min())

    orig = img.permute(1, 2, 0).cpu().numpy()

    plt.figure(figsize=(12, 4))
    plt.subplot(1, 3, 1)
    plt.imshow(np.clip(orig, 0, 1))
    plt.title(f"{cat} - Original")
    plt.axis("off")

    plt.subplot(1, 3, 2)
    plt.imshow(np.clip(recon.squeeze().permute(1, 2, 0).cpu().numpy(), 0, 1))
    plt.title("Reconstructed")
    plt.axis("off")

    plt.subplot(1, 3, 3)
    plt.imshow(diff_map, cmap='hot')
    plt.title("Anomaly Map")
    plt.axis("off")
    plt.show()

print("\n✅ All categories processed successfully!")
```
